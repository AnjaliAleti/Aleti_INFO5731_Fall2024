```python
# Write your solution below.
# Tip: the API response includes a metadata object plus job objects.
# Make sure you keep only real job postings.


import requests
import pandas as pd
from collections import Counter

API_URL = "https://remoteok.com/api"

# Send request (RemoteOK requires a user-agent header)
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(API_URL, headers=headers)

data = response.json()

# The first object is metadata — remove it
jobs = [job for job in data if isinstance(job, dict) and job.get("id")]

# Q1: Total number of job postings
total_jobs = len(jobs)
print("Q1: Total job postings =", total_jobs)

# Q2: Top 3 most frequent companies
companies = [job.get("company", "Unknown") for job in jobs]
top_3_companies = Counter(companies).most_common(3)
print("Q2: Top 3 companies =", top_3_companies)

# Q3: Titles containing "Data" (case-insensitive)
data_jobs_count = sum(
    1 for job in jobs
    if "data" in job.get("position", "").lower()
)
print('Q3: Jobs containing "Data" in title =', data_jobs_count)

# Q4: Create DataFrame with first 10 jobs
rows = []

for job in jobs[:10]:
    rows.append({
        "title": job.get("position"),
        "company": job.get("company"),
        "location": job.get("location") if job.get("location") else "Unknown",
        "salary": job.get("salary") if job.get("salary") else None,
        "tags": ", ".join(job.get("tags", []))
    })

df = pd.DataFrame(rows)

print("\nFirst 10 job postings:")
print(df.head(10))











```

Question 3 ( 2 points)
Create a program that accepts a list of numbers as input and outputs a new list containing only the even numbers.

```python
# write your answer here
# Program to filter even numbers from a list

# Take input from user (numbers separated by space)
numbers = input("Enter numbers separated by space: ")

# Convert input string into a list of integers
num_list = list(map(int, numbers.split()))

# Create a new list containing only even numbers
even_numbers = [num for num in num_list if num % 2 == 0]

# Display the result
print("Even numbers:", even_numbers)

```
