### Q3. Create a sentiment label from the compound score (3 points)
Create a new column called `PredictedSentiment` using this rule:
- `compound >= 0.05` → `positive`
- `compound <= -0.05` → `negative`
- otherwise → `neutral`

Then show the first 10 rows of `Text`, `compound`, and `PredictedSentiment`.

```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download VADER lexicon (only needed once)
nltk.download('vader_lexicon')

# Load dataset
df = pd.read_csv("sentiment_classification_combined_data.csv")

# Initialize VADER
sia = SentimentIntensityAnalyzer()

# Compute compound score
df['compound'] = df['Text'].astype(str).apply(lambda text: sia.polarity_scores(text)['compound'])

# Create PredictedSentiment column
def get_sentiment_label(score):
    if score >= 0.05:
        return 'positive'
    elif score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

df['PredictedSentiment'] = df['compound'].apply(get_sentiment_label)

# Show first 10 rows
print(df[['Text', 'compound', 'PredictedSentiment']].head(10))
```

### Q4. Compare review score and sentiment (2 points)
Group by `Score` and compute the **average compound score** for each score value.

```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download VADER lexicon (only needed once)
nltk.download('vader_lexicon')

# Load dataset
df = pd.read_csv("sentiment_classification_combined_data.csv")

# Initialize VADER
sia = SentimentIntensityAnalyzer()

# Compute compound score
df['compound'] = df['Text'].astype(str).apply(lambda text: sia.polarity_scores(text)['compound'])

# Group by Score and compute average compound score
avg_sentiment_by_score = df.groupby('Score')['compound'].mean()

# Display result
print(avg_sentiment_by_score)
```

### Q1. Zero-shot classification prompt (3 points)
Write a **zero-shot prompt** that asks an AI model to classify each review as **positive, negative, or neutral**.

Your prompt should include:
- a clear instruction
- the three labels
- a request for a clean output format

Store your prompt in a Python variable called `zero_shot_prompt`, then print it.

```python
import pandas as pd

# Load dataset
df = pd.read_csv("sentiment_classification_combined_data.csv")

# Take one sample review (you can change index if needed)
sample_text = df['Text'].astype(str).iloc[0]

# Create zero-shot prompt
zero_shot_prompt = f"""
You are an AI model that classifies customer reviews.

Instruction:
Classify the sentiment of the given review into one of the following labels:
- positive
- negative
- neutral

Review:
"{sample_text}"

Output Format:
Return only one word from [positive, negative, neutral].
"""

# Print the prompt
print(zero_shot_prompt)
```

### Q2. Few-shot prompt improvement (3 points)
Now improve the previous prompt by adding **two examples**.

Requirements:
- Store it in a variable called `few_shot_prompt`
- Include 2 input-output examples
- Ask the model to classify the 3 sample reviews after the examples

```python
import pandas as pd

# Load dataset
df = pd.read_csv("sentiment_classification_combined_data.csv")

# Select 3 sample reviews
samples = df['Text'].astype(str).head(3).tolist()

# Create few-shot prompt
few_shot_prompt = f"""
You are an AI model that classifies customer reviews into sentiment categories.

Instruction:
Classify each review as one of the following:
- positive
- negative
- neutral

Examples:
Review: "I absolutely love this product! It works perfectly."
Sentiment: positive

Review: "This is the worst purchase I have ever made."
Sentiment: negative

Now classify the following reviews:

Review 1: "{samples[0]}"
Review 2: "{samples[1]}"
Review 3: "{samples[2]}"

Output Format:
Return the answers in this format:
Review 1: <positive/negative/neutral>
Review 2: <positive/negative/neutral>
Review 3: <positive/negative/neutral>
"""

# Print the prompt
print(few_shot_prompt)
```

## Reflection (optional, no extra points)
Briefly explain the difference between:
- sentiment analysis
- text classification
- prompt engineering

Sentiment analysis focuses specifically on identifying the emotional tone of text (e.g., positive, negative, or neutral).
Text classification is more general; it classifies text according to any pre-established categories, including intent classification, subject labeling, and spam detection.
Prompt engineering is different from both; it involves designing and refining inputs (prompts) to guide a language model to produce better, more accurate outputs for tasks like classification, summarization, or reasoning.
